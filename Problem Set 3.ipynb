# Q1) Occupations

Step 1: Import the necessary libraries

```python
import pandas as pd
```

Step 2: Import Dataset from this address (https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user)

Step 3: Assign it to the variable called "USER"

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
```

Step 4: Discover the mean age per occupation

```python
#Method 1: Directly Referring to the colum name
meanage_occupation = users[['occupation','age']].groupby('occupation').mean()
meanage_occupation
```

```python
#Method2: Referring to column by position using "iloc"

meanage_occupation = users.iloc[:,[3,1]].groupby('occupation').mean()
meanage_occupation
```

Step 5: Discover the male ratio from the occupation and sort it from the most to the least

```python
dt = users[['gender','occupation']]
males = dt.loc[dt['gender']=='M']
groupby_occupations = dt.groupby('occupation').count()
malecountby_occupation = males.groupby('occupation').count()
malesratio_occupation = malecountby_occupation / groupby_occupations
malesratio_occupation.sort_values(ascending=False, by='gender')
```

Step 6: For each occupation. calculte the minimum and maximum ages

```python
dt = users[['occupation', 'age']]
occupationmin_age = dt.groupby('occupation').min()
occupationmax_age = dt.groupby('occupation').max()
results = pd.concat([occupationmax_age, occupationmin_age], axis=1)
results
```

Step 7: For each combination of occupation and sex, calculate the mean age

```python
dt = users[['occupation', 'gender', 'age']]
usermeanage_gender = dt.groupby(['occupation', 'gender']).mean('age')
usermeanage_gender
```

Step 8: For each occupation present the percentage of women and men

```python
gender_percentage = dt.groupby('occupation')['gender'].value_counts(normalize=True)*100
gender_percentage
```

Step 14. Downsample the record to a weekly frequency for each location.


```python
week = data.resample('W').ffill()

week
```

```python
weekstats = pd.concat([weekstats.mean(), weekstats.max(), weekstats.mean(), weekstats.std()], axis=1)
weekstats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)
weekstats
```
